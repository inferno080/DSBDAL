<font size="5">Imports</font>

```python
import pandas as pd
```

<font size="5">Read Dataset</font>

```python
df = pd.read_csv('Iris.csv')
df.head()
```

<font size="5">Data Preprocessing</font>

```python
df = df.dropna()
df.head()
```

<font size="5">Creating Subsets</font>

```python
subset_types = df["Species"].unique()
subset_types
```

```python
Iris_setosa_subset = df[df['Species'] == 'Iris-setosa']
Iris_versicolor_subset = df[df['Species'] == 'Iris-versicolor']
Iris_virginica_subset = df[df['Species'] == 'Iris-virginica']
```

```python
Iris_setosa_subset.head()
```

```python
Iris_virginica_subset.head()
```

<font size="5">Merging Subsets</font>

```python
merged_data = pd.concat([Iris_setosa_subset,Iris_virginica_subset], keys=["Species 1", "Species 3"])
merged_data
```

<font size="5">Sort Data</font>

```python
df.sort_values("SepalLengthCm").head()
```

<font size="5">Transposing Data</font>

```python
df_trasnpose = df.transpose()
df_trasnpose.head()
```

<font size="5">Melting Data</font>

```python
df_melted = df.melt()
df_melted.head()
```

<font size="5">Casting Data</font>

```python
df_casted = df.astype({"SepalLengthCm":"int32"})
df_casted.head()
```
