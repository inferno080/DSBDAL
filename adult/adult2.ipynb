<font size="5">Imports</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

<font size="5">Reading Dataset</font>

```python
df = pd.read_csv("adult.csv")
df.head()
```

<font size="5">Data Preprocessing</font>

```python
df = df.replace('?', np.NaN)
df = df.dropna()
df.head()
```

<font size="5">Visuzlizations</font>

```python
sns.histplot(df["age"])
```

```python
sns.scatterplot( df["hours-per-week"],df["age"])
```

```python
sns.barplot( df["gender"], df["hours-per-week"])
```

```python
sns.lineplot( df["age"],df["educational-num"])
```

```python
f, (ax_box, ax_hist, ax_scatter) = plt.subplots(3, sharex=True,  gridspec_kw={"height_ratios": (.2, .3, 0.5)})

sns.boxplot(df["age"], ax=ax_box)
sns.histplot(df["age"], ax=ax_hist)
sns.scatterplot(df["age"], df["hours-per-week"], ax=ax_scatter)
```

```python
countries = df['native-country'].value_counts().to_dict()
cdf = pd.DataFrame.from_dict([countries])
cdf = cdf.transpose()
cdf.columns =['Count']
cdf
```

```python
cdf.plot.pie(y='Count', figsize=(10,10))
```
