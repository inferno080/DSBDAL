<font size=5>Imports</font>

```python
import pandas as pd
import numpy as np
```

<font size=5>Reading Data</font>

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

<font size=5>Data Preprocessing</font>

```python
df = df.replace('?', np.NaN)
df = df.dropna()
df = df[df["release_year"]>0]
df.head()
```

<font size=5>Creating Subsets</font>

```python
subsets = df["type"].unique()
subsets
```

```python
Movies = df[df["type"] == "Movie"]
Series = df[df["type"] == "TV Show"]
```

```python
Movies.head()
```

```python
Series.head()
```

<font size=5>Merging Subsets</font>

```python
df_merged = pd.concat([Movies,Series], keys=["Movies", "Series"])
df_merged
```

<font size=5>Sorting Data</font>

```python
df = df.sort_values("release_year")
df.head()
```

<font size=5>Transposing Data</font>

```python
df_transpose = df.transpose()
df_transpose.head()
```

<font size=5>Melting Data</font>

```python
df_melted = df.melt()
df_melted.head()
```

<font size=5>Casting Data</font>

```python
df["release_year"] = df["release_year"].astype(float)
df.head()
```
